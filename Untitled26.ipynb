```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# Enable inline plotting in Jupyter
%matplotlib inline

# Load the dataset from the uploaded file
file_path = "/mnt/data/diabetes.csv"
pima_df = pd.read_csv("diabetes.csv")

# Display the first few rows to confirm column names
print(pima_df.head())

# If the first row contains the header, reload without specifying names
if pima_df.columns[0] == 'Pregnancies':
    colnames = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'target']
    pima_df.columns = colnames

# Check for non-numeric values in the dataset
non_numeric_rows = pima_df[~pima_df.applymap(np.isreal).all(1)]
print(non_numeric_rows)

# Replace missing values with the median of each column
pima_df = pima_df.fillna(pima_df.median())

# Display the distribution of the various attributes
print(pima_df.describe().transpose())

# Display the distribution of the target column
print(pima_df.groupby(["target"]).count())

# Define features (X) and target (Y)
X = pima_df.drop('target', axis=1)
Y = pima_df['target']

# Ensure all columns are numeric
X = X.apply(pd.to_numeric)
Y = Y.apply(pd.to_numeric)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=101)

# Fit the Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Make predictions
y_predict = model.predict(X_test)

# Evaluate the model
model_score = model.score(X_test, y_test)
print(f"Model Score: {model_score}")
print(f"Confusion Matrix:\n{metrics.confusion_matrix(y_test, y_predict)}")
print(f"Classification Report:\n{metrics.classification_report(y_test, y_predict)}")

# Plot the confusion matrix
cm = metrics.confusion_matrix(y_test, y_predict)
plt.clf()
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Wistia)
classNames = ['NonDiabetic', 'Diabetic']
plt.title('Confusion Matrix - Test Data')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)

# Annotate the confusion matrix with counts
for i in range(len(classNames)):
    for j in range(len(classNames)):
        plt.text(j, i, str(cm[i][j]), horizontalalignment='center', verticalalignment='center', color='black')

plt.show()
```

```python
import pickle

# Fit the Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Save the model to a pickle file
with open('logistic_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
import pickle

# Load the model from the pickle file
with open('logistic_regression_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

# Use the loaded model to make predictions
y_predict_loaded = loaded_model.predict(X_test)

# Evaluate the loaded model
model_score_loaded = loaded_model.score(X_test, y_test)
conf_matrix_loaded = metrics.confusion_matrix(y_test, y_predict_loaded)
class_report_loaded = metrics.classification_report(y_test, y_predict_loaded)

print(f"Loaded Model Score: {model_score_loaded}")
print(f"Loaded Confusion Matrix:\n{conf_matrix_loaded}")
print(f"Loaded Classification Report:\n{class_report_loaded}")
```
